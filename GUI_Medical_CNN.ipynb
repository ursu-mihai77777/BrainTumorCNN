```python


import gradio as gr
import tensorflow as tf
import numpy as np


try:
    MODEL = tf.keras.models.load_model(MODEL_PATH)
    print(" Keras model loaded successfully.")
except Exception as e:
    print(f" Error loading model: {e}")
    MODEL = None

def predict_brain_tumor(image):
    if MODEL is None:
        return {"Error": 1.0, "Details": "Model could not be loaded at startup."}

    try:

        img = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))


        img = tf.cast(img, dtype=tf.float32)


        img = img / 255.0


        img = np.expand_dims(img, axis=0)


        prediction = MODEL.predict(img)[0][0]


        prob_tumor = float(prediction)


        prob_no_tumor = 1.0 - prob_tumor


        results = {
            CLASS_NAMES[0]: prob_no_tumor,
            CLASS_NAMES[1]: prob_tumor
        }

        return results

    except Exception as e:
        print(f" ERROR DURING PREDICTION: {e}")
        return {"Processing Error": f"Error: {str(e)}"}


iface = gr.Interface(
    fn=predict_brain_tumor,
    inputs=gr.Image(width=IMG_SIZE, height=IMG_SIZE, label="Upload a Brain Tumor MRI Image"),
    outputs=gr.Label(num_top_classes=2, label="Classification Result"),
    title=" CNN Brain Tumor Detector",
    description="Upload an MRI image to get the model's prediction (No Tumor or Brain Tumor).",
    live=True
)


iface.launch(share=True)
```
